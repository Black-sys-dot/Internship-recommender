```python
# for each title, count how often each stipend shows up
counts = data.groupby(['Internship Title', 'Stipend'])['Stipend'].transform('count')

# keep only rows where stipend is unique inside its title group
filtered = data[counts == 1]

print(filtered.head(20))  # check first 20 rows
```

```python
import re
import numpy as np
def clean_payroll(x):
    if pd.isna(x):
        return np.nan
    x_clean = x.replace(',', '').strip().lower()
    
    if x_clean in ['unpaid', 'negotiable']:
        return 0  # treat as zero, or keep np.nan if you prefer
    
    if '-' in x_clean:
        nums = re.findall(r'\d+', x_clean)
        return np.mean([int(n) for n in nums]) if nums else np.nan
    else:
        num = re.search(r'\d+', x_clean)
        return int(num.group()) if num else np.nan

filtered['Payroll_Clean'] = filtered['Stipend'].apply(clean_payroll)
```

```python
filtered
```

```python
def assign_levels(group, num_levels=5):
    group = group.copy()
    
    # Drop NaNs to avoid messing up qcut
    payroll = group['Payroll_Clean'].dropna()
    
    if payroll.nunique() < 2:
        # If all values are the same → only one level possible
        group['Level'] = "Single Level"
    else:
        try:
            # Try to cut into quantiles
            categories = pd.qcut(
                payroll,
                q=num_levels,
                labels=[
                    'Level 1 - Elite',
                    'Level 2 - Senior',
                    'Level 3 - Mid',
                    'Level 4 - Junior',
                    'Level 5 - Entry'
                ][:num_levels],  # slice so labels match bins
                duplicates="drop"
            )
            
            # Assign back to group
            group['Level'] = categories.reindex(group.index)  # align indices
        except ValueError:
            # Fallback if qcut still fails
            group['Level'] = "Uncategorized"
    
    return group

# Apply per internship title
filtered = filtered.groupby('Internship Title', group_keys=False).apply(assign_levels)
```

```python
import ast

unique_skills = set()

for skills in df["Skills_List"]:
    if isinstance(skills, str):
        try:
            # Case 1: it's actually a string that looks like a list
            parsed = ast.literal_eval(skills)
            if isinstance(parsed, (list, tuple)):
                unique_skills.update(s.strip() for s in parsed)
            else:
                # Case 2: just a comma-separated string
                unique_skills.update(s.strip() for s in skills.split(","))
        except (SyntaxError, ValueError):
            # fallback: comma-separated text
            unique_skills.update(s.strip() for s in skills.split(","))
    elif isinstance(skills, (list, tuple)):
        unique_skills.update(s.strip() for s in skills if isinstance(s, str))

print(unique_skills)
```

```python
import torch
import torch.nn as nn
import ast
device=torch.device("cuda")
# Step 1: skill-to-index mapping
skill2idx = {skill: i for i, skill in enumerate(sorted(unique_skills))}

embedding_dim = 32  # each skill is a 32-dim vector
embedding_matrix = nn.Parameter(
    torch.randn(len(unique_skills), embedding_dim, device=device)
)

df["Skills_List"] = df["Skills_List"].fillna("[]")
df["Skills_List"] = df["Skills_List"].apply(ast.literal_eval)

# Step 2: find max number of skills in any row
max_len = df["Skills_List"].apply(len).max()
print("Longest Skills_List length:", max_len)

# Step 3: build row_embeddings as a list of PyTorch tensors
row_embeddings = []

for skills in df["Skills_List"]:
    vecs = []
    if isinstance(skills, (list, tuple)):
        for skill in skills:
            if skill in skill2idx:
                idx = skill2idx[skill]
                vecs.append(embedding_matrix[idx])  # this is a tensor with grad

    if vecs:
        mat = torch.stack(vecs, dim=0)  # shape: (num_skills, embedding_dim)
    else:
        mat = torch.zeros((0, embedding_dim), device=device)

    # pad to max_len
    if mat.shape[0] < max_len:
        pad = torch.zeros((max_len - mat.shape[0], embedding_dim), device=device)
        mat = torch.cat([mat, pad], dim=0)

    row_embeddings.append(mat)

# Step 4: optionally stack to get a single 3D tensor (num_rows x max_len x embedding_dim)
row_embeddings = torch.stack(row_embeddings, dim=0)  # shape: (num_rows, max_len, embedding_dim)

print("row_embeddings shape:", row_embeddings.shape)
```

```python
# This shit seems ok
import torch
from torch.utils.data import Dataset, DataLoader
import random
import torch.nn as nn
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class TripletDataset(Dataset):
    def __init__(self, main_frame, row_embeds, level_embeds, skills_list):
        if isinstance(main_frame, torch.Tensor):
            main_frame = main_frame.cpu().numpy()
        self.main_frame = main_frame


        self.row_embeds = row_embeds if torch.is_tensor(row_embeds) else torch.tensor(row_embeds)
        self.level_embeds = level_embeds if torch.is_tensor(level_embeds) else torch.tensor(level_embeds)
        self.row_embeds=self.row_embeds.to(device)
        self.level_embeds=self.level_embeds.to(device)
        self.skills_list = skills_list

        self.positives = {}
        self.negatives = {}
        self.valid_indices = []

        for i in range(len(main_frame)):
            pos, neg = [], []
            for j in range(len(main_frame)):
                if i == j:
                    continue
                label = self._check_rules(i, j)
                if label == "positive":
                    pos.append(j)
                else:
                    neg.append(j)
            if pos and neg:
                self.positives[i] = pos
                self.negatives[i] = neg
                self.valid_indices.append(i)

    def _check_rules(self, anchor_idx, pair_idx):
        a_level, a_loc, a_title = self.main_frame[anchor_idx]
        p_level, p_loc, p_title = self.main_frame[pair_idx]


        a_level, p_level = int(a_level), int(p_level)
        a_skills = set(self.skills_list[anchor_idx])
        p_skills = set(self.skills_list[pair_idx])

        if (a_title != p_title) or (a_loc != p_loc):
            return "negative"

        if a_skills < p_skills:
            return "negative"

        if p_skills.issubset(a_skills):
            if p_level <= a_level:
                return "positive"
            else:
                return "negative"

        return "negative"

    def _build_vector(self, i, mode="anchor"):
        frame = torch.tensor(self.main_frame[i,1:], dtype=torch.float32).flatten().to(device)
        if mode == "anchor":
            # anchor: frame + row_embeddings
            row = self.row_embeds[i].float().flatten()
            return torch.cat([frame, row], dim=0)
        else:
            # pos/neg: frame + level_embeddings
            level = self.level_embeds[i].float().flatten()
            return torch.cat([frame, level], dim=0)

    def __getitem__(self, idx):
        anchor_idx = self.valid_indices[idx]
        pos_idx = random.choice(self.positives[anchor_idx])
        neg_idx = random.choice(self.negatives[anchor_idx])
        return (
            self._build_vector(anchor_idx, mode="anchor"),
            self._build_vector(pos_idx, mode="pos"),
            self._build_vector(neg_idx, mode="neg"),
        )

    def __len__(self):
        return len(self.valid_indices)


# ------------------------
# Triplet Neural Network
# ------------------------
class TripletNN(nn.Module):
    def __init__(self, input_dim, hidden_dims=[512, 512, 512], output_dim=256):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dims[0])
        self.fc2 = nn.Linear(hidden_dims[0], hidden_dims[1])
        self.fc3 = nn.Linear(hidden_dims[1], hidden_dims[2])
        self.fc4 = nn.Linear(hidden_dims[2], output_dim)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        x = self.fc4(x)
        return x


# ------------------------
# Example Usage
# ------------------------
if True:
    #main_frame_arr = np.array(main_frame_arr, dtype=int)  # force numpy ints
    dataset = TripletDataset(main_frame_arr, row_embeddings, level_embeddings, df['Skills_List'])
    dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

    all_batches=list(dataloader)
    train_batches=all_batches[:-2]
    test_batches=all_batches[-2:]

    # Now input = main_frame(3) + either row(256) or level(256)
    input_dim = 2 + 8*32  

    model = TripletNN(input_dim=input_dim).to(device)
    criterion = nn.TripletMarginLoss(margin=1.0, p=2)
    optimizer = optim.Adam(
    list(model.parameters()) + [level_embedding_matrix,embedding_matrix],
    lr=1e-5
    )

# --- Training Loop ---
num_epochs = 10
for epoch in range(num_epochs):
    total_loss = 0.0
    j=0
    for anchor_vec, pos_vec, neg_vec in train_batches:
        anchor_vec = anchor_vec.to(device)
        pos_vec = pos_vec.to(device)
        neg_vec = neg_vec.to(device)

        optimizer.zero_grad()
        anchor_emb = model(anchor_vec)
        pos_emb = model(pos_vec)
        neg_emb = model(neg_vec)

        loss = criterion(anchor_emb, pos_emb, neg_emb)
        loss.backward()
        optimizer.step()
        j+=1

        total_loss += loss.item()
        print(f"{total_loss}",end="\r")

    print(f"Epoch {epoch+1}/{num_epochs} | Loss: {total_loss/len(dataloader):.4f}")
```
